```python
#import libraries
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
from sklearn import cluster, datasets, metrics

kmeans2015 = [] #create a blank list
ks = range(2, 11) #given range for clusters
for k in ks: #use for loop to run all numbers of clusters
    kmeans_fit = KMeans(n_clusters = k).fit(X) 
    cluster_labels = kmeans_fit.labels_
    kmean = metrics.silhouette_score(X, cluster_labels)
    kmeans2015.append(kmean) #append number into list

plt.bar(ks, kmeans2015) #plot the outcome
plt.title('Results of Different Clusters - K-Means') #named title
plt.xlabel('Cluster group')
plt.ylabel('Count')
plt.savefig('Figure 4.2.png', bbox_inches='tight') #output the plot
plt.show()
print(kmeans2015) #print the numerical data
```

```python
#create def function for kmeans
def doKmeans(X, nclust=2): #set number for cluster
    model = KMeans(nclust) #run kmeans
    model.fit(X) #fit data
    clust_labels = model.predict(X) #predict data
    cent = model.cluster_centers_ #calculate central point
    return (clust_labels, cent) #return value

clust_labels, cent = doKmeans(love2015_2, 2) #assign value
kmeans = pd.DataFrame(clust_labels) #turn into dataframe
love2015_2.insert((love2015_2.shape[1]),'kmeans',kmeans) #insert value
```
